### OLS estimates
We are given dataset of 6 variables (2 independent, 4 dependent) with 11 observations. We are to give an OLS and MLE estimate of $\beta$ coefficient for 4 linear models:
- $Y_1 = \beta_{0,1} + \beta_{1,1} X_1 + \epsilon_1$
- $Y_2 = \beta_{0,2} + \beta_{1,2} X_1 + \epsilon_2$
- $Y_3 = \beta_{0,3} + \beta_{1,3} X_1 + \epsilon_3$
- $Y_4 = \beta_{0,4} + \beta_{1,4} X_2 + \epsilon_4$

Denote $X_j^i$ and $Y_j^i$ the $i$-th observation of random sample $X_j$, $Y_j$ respectively.

We start by loading and viewing the data.

```r
(x <- data.frame(read.csv("data/hw1_data.csv"))) # load the data
```

Both independent variables $X_1^i$ and $X_2^i$ seems to come from discrete distribution, the dependent variables $Y_1^i,\dots,Y_4^i$ seems to come from continuous distribution. 

Firstly, it is worth noting that random sample $X_2$ contains almost no useful information (at least for estimation of parameters in last linear model) as there is only one observation that is different from the other ones. This also relates with the second comment - the number of observation is not exactly large. Therefore one has to be careful, when deriving any conclusions or implications of the results.

We provide scater plots of the mentioned touples of variables to get a visual intuition of the (in)dependency.

```r
# Providing scatter plots for each pair
attach(x)
par(mfrow=c(2,2))
plot(X1, Y1, main="Scatter Plot - Model 1", xlab="X_1", ylab="Y_1", pch=19)
plot(X1, Y2, main="Scatter Plot - Model 2", xlab="X_1", ylab="Y_2", pch=19)
plot(X1, Y3, main="Scatter Plot - Model 3", xlab="X_1", ylab="Y_3", pch=19)
plot(X2, Y4, main="Scatter Plot - Model 4", xlab="X_2", ylab="Y_4", pch=19)
```

### MLE estimates

We proceed by MLE estimation of $\beta$-coefficients of the previous stated 4 models. We use following derived formulas to compute them: 
$$
\begin{align}
\hat \beta_1 &= \frac{\sum\limits_{j=1}^{11} (x_i - \overline x)(y_i - \overline y)}{\sum\limits_{j=1}^{11} (x_i - \overline x)^2}, \\
\\
\hat \beta_0 &= \overline y - \hat \beta_1 \overline x.
\end{align}$$

```r
# Functions for MLE computation of beta-coefficients

# MLE estimate for beta_1
beta1 = function(x,y){
    nahore = sum((x - mean(x))*(y-mean(y))) # citatel
    dole = sum((x - mean(x))*(x-mean(x))) # jmenovatel
    b1 = nahore / dole # vysledna beta
    return(b1)
}
# MLE estimate for beta_0
beta0 = function(b1,x,y){
    b0 = mean(y)-b1*mean(x) # vysledna beta
    return(b0)
}
```

```r
# betas - model 1
b11 = beta1(X1,Y1)
b10 = beta0(b11,X1,Y1)

# betas - model 2
b21 = beta1(X1,Y2)
b20 = beta0(b21,X1,Y2)

# betas - model 3
b31 = beta1(X1,Y3)
b30 = beta0(b31,X1,Y3)

# betas - model 4
b41 = beta1(X2,Y4)
b40 = beta0(b41,X2,Y4)

# print out the results
betas <- matrix(c(b10,t(lm1[1][[1]])[1],b11,t(lm1[1][[1]])[2],b20,t(lm2[1][[1]])[1],b21,t(lm2[1][[1]])[2],b30,t(lm3[1][[1]])[1],b31,t(lm3[1][[1]])[2],b40,t(lm4[1][[1]])[1],b41,t(lm4[1][[1]])[2]),ncol=4,byrow=TRUE)
colnames(betas) <- c("Beta_0 - MLE","Beta_0 - OLS","Beta_1 - MLE", "Beta_1 - OLS")
rownames(betas) <- c("Model 1","Model 2","Model 3","Model 4")
data.frame(betas)
 
```

As we can see the estimations of $\beta$ are both same for MLE and OLS for each model.

We interpret the diagnostic plots. 
#### Residuals vs. fitted plots
Firstly we comment on each residuals vs. fitted plots. If assumptions of OLS model hold the points in this plot should be distributed independently $N(0,\sigma^2)$ around $0$ on $y$-axis, i.e. we should not see any trend (curvature) or changing variance in distribution of residuals. $\textbf{Model 1}$ seems not to violate this rule. $\textbf{Model 2}$ on the other hand violates this assumption as there is clear dependence between regressors and residuals. We should therefore adjust our model as there is still some information in regressors which is unused. As mentioned above, some non-linear transformation (probably adding $X_1^2$ to our established model) of regressors might help to get better result. $\textbf{Model 3}$ has a downward trend because the one mentioned outlier changed dramatically the slope of the regression line. $\textbf{Model 4}$ (as mentioned above) would need more observation on level $X_2 = 19$. In our scenario it his hard to judge whether the model satisfies our visual assumptions about residuals vs. fitted plot, because we have no information about variance on the level $X_1 = 19$.

#### Q-Q plots of residuals
Q-Q plots are meant to reveal if residuals are (more or less) normally distributed. It compares the empirical quantiles of residuals and quantiles of normal distribution. In the case of $\textbf{model 1}$ it seems the residuals might come from another than normal distribution, but we have to keep in mind that we have only $11$ observation, therefore the discrepancies might disappear if we had more data. In Q-Q plot of $\textbf{model 2}$ we can see grouped touples of points which are of the same distance from 0 which can be seen in Residuals vs. fitted plot. Q-Q plot of $\textbf{model 3}$ has a slightly narrower slope than the line corersponding to normal distribution - this is again caused by one outlier which causes more residuals with negative sign. For $\textbf{model 4}$ we apply the similar reasoning as in the case of model 1 - we would need more data to confirm whether the residuals are normally distributed. Although in this model we can see less discrepancy than in the model 1.

#### Scale-location
Scale-location plot is useful for checking the assumption of homoscedasticity. If the assumptions of OLS hold we shouldn't be able to see any pattern/trend in the data. The points should be somehow randomly distributed across the straight line with a slope equal to $0$. In the $\textbf{model 1}$ there is no great sign of heteroscedasticity. In the $\textbf{model 2}$ however we can again see that the linear fit is not the best one. Due to a above mentioned curvature the variance will grow for lower and higher values. $\textbf{Model 3}$ again doesn't give desired results due to one outlier. We can see that variance increases with higher values. For the $\textbf{model 4}$ there is no meaningful outcome of a Scale-location plot, as there is only one level for which we can estimate variance.

#### Residuals vs. leverage
This last diagnostic plot is meant to give us influential points or outliers. We are primarily interested in values which appear in the upper right or upper left corner. In the $\textbf{model 1}$ the most influential point is $(13,7.58)$. If we look at the scatter plot with fitted line this makes sense as it slightly decreases slope of the regression line. In the $\textbf{model 2}$ the most influential point is $(4,3.10)$. It is possible that without this point the regression line would fit the dataset much better, but as mentioned above this model has different flaws which should be handled by improving the regression function. In the $\textbf{model 3}$ we can clearly see what was evident by visual approach at a first glance. The point $(12,13.74)$ sharply changed the slope of otherwise almost-in-one-line point pattern. For $\textbf{model 4}$ the absence of influential points basically means that on the level $X_2 = 8$, there is no great outlier.

### Quantile regression estimates


#### LAD estimates

We move to the LAD estimate which gives a conditional median (not conditional mean as above). This method is more robust and more resistant to outliers then previous OLS method. We provide a comparison of OLS and LAD of each model considered.

```r
library('quantreg')
s11lad <- rq(Y1 ~ X1, data = x, 0.5)
s12lad <- rq(Y2 ~ X1, data = x, 0.5)
s13lad <- rq(Y3 ~ X1, data = x, 0.5)
s24lad <- rq(Y4 ~ X2, data = x, 0.5) # solution may be non-unique
```

```r
library('stargazer')
stargazer(lm1, lm2, lm3, lm4, s11lad, s12lad, s13lad, s24lad,
          type="text", covariate.labels = c("X1","X2", "Intercept"), 
          column.labels = c("s11ols", "s12ols", "s13ols", "s24ols", "s11lad", "s12lad", "s13lad", "s24lad"))
```

```r
par(mfrow = c(2,2))
plot(Y1 ~ X1, data = x, main = "OLS (red) and LAD (blue) - Model 1")
abline(lm1, col = "red")
abline(s11lad, col = "blue")

plot(Y2 ~ X1, data = x, main = "OLS (red) and LAD (blue) - Model 2")
abline(lm2, col = "red")
abline(s12lad, col = "blue")

plot(Y3 ~ X1, data = x, main = "OLS (red) and LAD (blue) - Model 3")
abline(lm3, col = "red")
abline(s13lad, col = "blue")

plot(Y4 ~ X2, data = x, main = "OLS (red) and LAD (blue) - Model 4")
abline(lm2, col = "red")
abline(s24lad, col = "blue")
```

As we can see the LAD fits perfectly the model 3 as median is generally resistant to outliers which was the main issue in an OLS estimation of this model. In the model 1 there is a slight change of slope and intercept corresponding with one influential point commented above. Model 2 has the same slope, but the intersect moved upwards. Similarly as in model 1 it corresponds with lowering the weight on the error of influential point. In the model 4 the change corresponds with a slight difference of sample median and sample mean on the level $X_2 = 8$.
